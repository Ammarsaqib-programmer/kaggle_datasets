```python
sns.set(style="whitegrid")

df['Addiction_Level_Rounded'] = df['Addiction_Level'].round().astype(int)

plt.figure(figsize=(12, 6))
ax = sns.countplot(data=df, x='Addiction_Level_Rounded', hue='Gender', palette='pastel')

plt.title('Count of Children by Addiction Level and Gender')
plt.xlabel('Screen Addiction Level (1â€“10)')
plt.ylabel('Number of Individuals')

for p in ax.patches:
    count = int(p.get_height())
    if count > 0:
        ax.annotate(count,
                    (p.get_x() + p.get_width() / 2., count),
                    ha='center', va='bottom',
                    fontsize=9, color='black')

plt.tight_layout()
plt.show()
```

```python
top_loc= df.groupby('Location')['Addiction_Level'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6)) 
top_loc= top_loc.reset_index()
sns.barplot(x='Addiction_Level', y='Location', data=top_loc, palette='viridis')
plt.title('Top 10 Locations by Average Addiction Level')
plt.xlabel('Average Addiction Level')
plt.ylabel('Location')
plt.show()
```

```python
top_anx = df.groupby("Anxiety_Level")["Age"].mean().sort_values(ascending=False)
top_anx = top_anx.reset_index()
sns.barplot(data=top_anx, x='Anxiety_Level', y='Age', palette='rocket')

plt.title('Most Anxious Age ')
plt.xlabel('Age')
plt.ylabel('Average Anxiety Level')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
